# SECTION 3 — NHANES DATASET

## 3.1 Load your merged NHANES dataset

```python
import pandas as pd

def load_cycle(suffix):
    demo = pd.read_sas(f"../data/nhanes_raw/DEMO_{suffix}.XPT")
    bmx  = pd.read_sas(f"../data/nhanes_raw/BMX_{suffix}.XPT")
    hba1c = pd.read_sas(f"../data/nhanes_raw/GHB_{suffix}.XPT")
    glu   = pd.read_sas(f"../data/nhanes_raw/GLU_{suffix}.XPT")
    diq   = pd.read_sas(f"../data/nhanes_raw/DIQ_{suffix}.XPT")

    df = (demo.merge(bmx,  on="SEQN", how="inner")
               .merge(hba1c, on="SEQN", how="inner")
               .merge(glu,   on="SEQN", how="inner")
               .merge(diq,   on="SEQN", how="inner"))
    
    df["cycle"] = suffix   # track which cycle it came from
    return df

# choose which cycles you want
cycles = ["G", "H", "I", "J"]   # 2011–2018

df = pd.concat([load_cycle(c) for c in cycles], ignore_index=True)
print(df.shape)


# use the merged multi-cycle data as "nhanes"
nhanes = df.copy()

print("DIQ010 in columns:", "DIQ010" in nhanes.columns)

df.head()
```

## 3.2 Select relevant columns

```python
columns_to_keep = [
    "SEQN",
    "RIAGENDR",     # gender
    "RIDAGEYR",     # age
    "RIDRETH1",     # race/ethnicity
    "INDFMPIR",     # income-to-poverty ratio
    "BMXBMI",       # BMI
    "LBXGH",        # HbA1c
    "LBXGLU",       # fasting glucose
    "DIQ010"        # diabetes diagnosis (target)
]

nhanes_sub = nhanes[columns_to_keep].copy()
nhanes_sub.head()
```

## 3.3 CLEAN THE TARGET LABEL

```python
# DIQ010 meaning:
# 1 = diabetic, 2 = no, 3 = borderline, 7/9 = refused/don't know

nhanes_sub = nhanes_sub[nhanes_sub["DIQ010"].isin([1,2,3])]

nhanes_sub["diabetes_label"] = nhanes_sub["DIQ010"].replace({
    1: 1,   # Yes → diabetic
    2: 0,   # No → non-diabetic
    3: 0    # Borderline → treat as non-diabetic
})

nhanes_sub.drop(columns=["DIQ010"], inplace=True)
```

## 3.4 HANDLE MISSING VALUES

```python
# These columns are essential and should not be missing
required_cols = ["RIDAGEYR", "INDFMPIR", "BMXBMI", "LBXGH", "LBXGLU"]
nhanes_sub = nhanes_sub.dropna(subset=required_cols)
```

## 3.5 ENCODE CATEGORICAL VARIABLES

```python
# Gender: 1 = male → 0, 2 = female → 1
nhanes_sub["RIAGENDR"] = nhanes_sub["RIAGENDR"].replace({1:0, 2:1})

# Race/ethnicity one-hot encoding
nhanes_sub = pd.get_dummies(
    nhanes_sub,
    columns=["RIDRETH1"],
    prefix="race",
    drop_first=True
)
```
